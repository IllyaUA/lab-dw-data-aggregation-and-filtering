## Countplots

As we can see, the company has more female customers than males. However, remember the old said:

> An image better than a thousand words.

Let's see how to plot this information. Before doing anything, we need to consider what its the right plot to graph this information. In case, we want to show frequencies. Therefore and bar plot can be the ideal candidate. 

Let's see how to create a [barplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html) first with matplotlib and the with seaborn to see the differences.

It's important to create a pair of `fig` and `ax` before anything.
In this case, we're going to create a canvas with only one plot. Therefore, ax will be a single variable. In contrast if we want to create several plots, ax will be a Numpy array.

```python
fig, ax = plt.subplots()
ax.bar(x = data['gender'].value_counts().index, height = data['gender'].value_counts().values)
plt.show()
```

However, this plot is very basic and it doesn't have axis labels. Let's see how to add them.

```python
fig, ax = plt.subplots()
ax.bar(x = data['gender'].value_counts().index, height = data['gender'].value_counts().values)
ax.set_xlabel("Gender")
ax.set_ylabel("Count")
plt.show()
```

Now, we can add a title.

```python
fig, ax = plt.subplots()
ax.bar(x = data['gender'].value_counts().index, height = data['gender'].value_counts().values)
ax.set_xlabel("Gender")
ax.set_ylabel("Count")
ax.set_title("Amount of customers of each gender")
plt.show()
```

Finally it's important to emphasize in some way what we want to highlight making use os strategic colors. In this case, we might be intereseted in highligting the bigger frequency of females. Then, one option is to color differently the bar for females. Let's see how we can control the color.

```python
fig, ax = plt.subplots()
ax.bar(x = data['gender'].value_counts().index, height = data['gender'].value_counts().values, color = ['steelblue', 'grey'])
ax.set_xlabel("Gender")
ax.set_ylabel("Count")
ax.set_title("Amount of customers of each gender")
plt.show()
```

with this litle change, we're automatically sending the audience view to the female bar. To get a list of colors available in matplotlib, check [here](https://matplotlib.org/stable/gallery/color/named_colors.html)

Now let's do the same using seaborn.

```python
fig, ax = plt.subplots()
sns.countplot(data=data, x="gender", ax = ax)
plt.show()
```

Several differences can be observed. First, the seaborn function to plot frequencies has a different name `.countplot()`. Second, seaborn automatically adds axis labels. Third, the bars are colored according to the gender with default colors. Finally, to create plots ins seaborn, usually we need to have the data stored in a Pandas DataFrame (with exceptions that will be covered later during the bootcamp)

We can also change the axis labels, and the title in seaborn.

```python
fig, ax = plt.subplots()
sns.countplot(data=data, x="gender", ax = ax)
ax.set_xlabel("Gender")
ax.set_ylabel("Count")
ax.set_title("Amount of customers of each gender")
plt.show()
```

We can also change in which order we want the bars to be using the option `order`.

```python
fig, ax = plt.subplots()
sns.countplot(data=data, x="gender", order=['F','M'], ax = ax)
ax.set_xlabel("Gender")
ax.set_ylabel("Count")
ax.set_title("Amount of customers of each gender")
plt.show()
```

Finally we can change the bar colors with the option `pallete`.

```python
fig, ax = plt.subplots()
sns.countplot(data=data, x="gender", order=['F','M'], palette=['steelblue', 'grey'], ax = ax)
ax.set_xlabel("Gender")
ax.set_ylabel("Count")
ax.set_title("Amount of customers of each gender")
plt.show()
```

Now let's see how we can plot the matplotlib, and seaborn plots together side by side. To do this we need to pass two integer numbers to the `plt.subplots()` function. The first one is how many rows we want in the canvas, while the second is how many columns. In our case, we want the plots arranged side by side. Therefore, we want one row and two columns. Thus our values will be 1,2.

On the other had, now we will have two plots. Therefore, the `ax` variable will become a Numpy array, and we will use it to specify where we want to add the corresponding plot:

ax[0] -> first plot
ax[1] -> second plot

```python
fig, ax = plt.subplots(1,2)

ax[0].bar(x = data['gender'].value_counts().index, height = data['gender'].value_counts().values, color = ['steelblue', 'grey'])
ax[0].set_xlabel("Gender")
ax[0].set_ylabel("Count")
ax[0].set_title("Amount of customers of each gender")

sns.countplot(data=data, x="gender", order=['F','M'], palette=['steelblue', 'grey'], ax = ax[1])
ax[1].set_xlabel("Gender")
ax[1].set_ylabel("Count")
ax[1].set_title("Amount of customers of each gender")

plt.show()
```

Unfortunatelly, both plots are overlapping. We can avoid this adding the option `plr.tight_layout()` right before the `plt.show()`

```python
fig, ax = plt.subplots(1,2)

ax[0].bar(x = data['gender'].value_counts().index, height = data['gender'].value_counts().values, color = ['steelblue', 'grey'])
ax[0].set_xlabel("Gender")
ax[0].set_ylabel("Count")
ax[0].set_title("Amount of customers of each gender")

sns.countplot(data=data, x="gender", order=['F','M'], palette=['steelblue', 'grey'], ax = ax[1])
ax[1].set_xlabel("Gender")
ax[1].set_ylabel("Count")
ax[1].set_title("Amount of customers of each gender")

plt.tight_layout()
plt.show()
```

Even with this option, both plots are too close. We have another option to prevent problem setting the size of each plot inside `plt.subplots()` using the option `fig_size=(xsize, ysize)`

where `xsize` and `ysize` are integer number in inches.

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

ax[0].bar(x = data['gender'].value_counts().index, height = data['gender'].value_counts().values, color = ['steelblue', 'grey'])
ax[0].set_xlabel("Gender")
ax[0].set_ylabel("Count")
ax[0].set_title("Amount of customers of each gender")

sns.countplot(data=data, x="gender", order=['F','M'], palette=['steelblue', 'grey'], ax = ax[1])
ax[1].set_xlabel("Gender")
ax[1].set_ylabel("Count")
ax[1].set_title("Amount of customers of each gender")

plt.tight_layout()
plt.show()
```

## Boxplots

There are other kind of plots that can be interesting to do: [boxplot](https://www.simplypsychology.org/boxplots.html).
This kind plot is a graphical way to see the quantiles of a numerical column. In particular, it shows the 25%, 50% 75% and some additional elements called "wiskers" that give us an idea of where most of the values are contained. Extreme values of the column will be far away from the "wiskers".

Let's see how we can create them with matplotlib and seaborn

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].boxplot(data['median_home_val'])
ax[1].boxplot(data['median_home_val'], vert=False)
plt.show()
```

In the previous plot, we can see a lot of points beyond one of the `wiskers`. This is a clear indication that we have several customers with big values of `median_home_val`.

Now let's do the same plots using seaborn.

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
sns.boxplot(data, y='median_home_val', ax=ax[0])
sns.boxplot(data, x='median_home_val', ax=ax[1])
plt.show()
```

Again, seaborn added the axis labels by default. In addition, the boxes are colored automatically and are wider. However, the information is the same.

In addition, we can create different boxplots for each value of a categorical (non-numerical) column. Let's see how to do this in seaborn.

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
sns.boxplot(data, y='median_home_val', x='gender', ax=ax[0])
sns.boxplot(data, x='median_home_val', y='gender', ax=ax[1])
plt.show()
```

Apparently, there are no clear differences. However, to be more precise we will have to conduct an statistical test to evaluate quantitativelly if this is True or False. This is something that will be coverd later during the Boot Camp.

## Barplots

Earlier, we have seen how we can do aggregations according to categorical columns. Let's compute what is the average `median_home_values` by gender. Then we will display this information unsing a bar plot

```python
group_by_gender = data.groupby('gender').agg({'median_home_val': 'mean'}).round(2)
group_by_gender
```

We can plot this using matplotlib and a barplot. However, it's more convenient to use seaborn, as we don't need to perform first the aggregation. Seaborn will do it for us.

Keep in mind that seaborn's `.countplot()` and `.barplot()` functions are not the same. The first one is used to displey frequencies, while the second to display averages.

```python
fig, ax = plt.subplots()
sns.barplot(data=data, x="gender", y="median_home_val", ax = ax)
plt.show()
```

The tiny black bars represents the [standard deviation](https://www.nlm.nih.gov/nichsr/stats_tutorial/section2/mod8_sd.html#:~:text=A%20standard%20deviation%20(or%20%CF%83,data%20are%20more%20spread%20out.). The longer the line, the bigger the standard deviation.

## Histograms and distributions

We have seen how to plot frequencies of two categorical columns. There is another type of plot to plot frequencies for numerical values. You might say, that this is almost impossible because if we consider float numbers because it's very inlikely to have twice the same float number of a column. Yes, that's True! But what we can do is to create bins that cover ranges of values and display how many float number we have in a given range. We can do this with matplotlib and seaborn. Let's see some examples.

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].hist(data["median_home_val"], bins=50)
sns.histplot(data, x = "median_home_val", bins=50, ax=ax[1])
plt.show()
```

## Scatterplots

Sometimes, we are interested to visually check if there are **linear relationships** between the two numerical columns. The best way to check this is using an [scatterplot](https://chartio.com/learn/charts/what-is-a-scatter-plot/). In a scatter plot, what we do is to represent each line of our DataFrame with a marker. Therefore, we will have as many markers as rows in our DataFrame. The location of each marker is given by the values of each column 

```python
fix, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].scatter(x=data['ic2'],y=data['ic3'])
ax[0].set_xlabel("IC2")
ax[0].set_ylabel('IC3')
sns.scatterplot(data, x='ic2', y='ic3', ax=ax[1])
#sns.scatterplot(data=data,x='ic2',y='ic3', hue='gender')
#sns.scatterplot(x=data['ic2'], y=data['ic3'])
plt.tight_layout()
plt.show()
```

It looks that there is a linear relationship except for one point. We would like to know who is this customer. However, those plots are not interactive. We will need another plotting library to be able to `hover` over the marker to know who is this customer. 
A cool feature of the seaborn's function is that aloows us to color the markers according to one categorical column. Let's do it for `gender`


```python
fix, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].scatter(x=data['ic2'],y=data['ic3'])
ax[0].set_xlabel("IC2")
ax[0].set_ylabel('IC3')
sns.scatterplot(data, x='ic2', y='ic3', hue='gender',ax=ax[1])
plt.tight_layout()
plt.show()
```

We could think of doing this kind of plots for all the pairs of numerical columns in our Pandas DataFrame. However, this will be a bit painful and seaborn  has an speciall function to do it.

```python
sns.pairplot(data.select_dtypes(np.number))
plt.show()
```

As shown, this produces a grid of plots with all the possible combinations of scatterplots of two numerical columns outside the diagonal. In contrast, in the diagonal, is showing the histogram of each numerical column

1. Create a Python function that takes a Pandas DataFrame, and two column names and creates an scatterplot (using seaborn) of them. 
Add an option to color the markers according to another categorical column. Name your function as `scatter_plot`

Remember to add the docstring and type hinting!!!

Hint: the type for strings is "str" and this function should not return anything.

```python
%%writefile snsscatter.py
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from typing import Optional, Tuple

def scatter_plot(df: pd.DataFrame, x_col: str, y_col: str, hue_col: Optional[str] = None) -> Tuple[plt.Figure, plt.Axes]:
    '''
    Create a scatterplot of two columns from a Pandas DataFrame.

    Parameters:
    - df: Pandas DataFrame
        The DataFrame containing the data to be plotted.
    - x_col: str
        The name of the column to be used as the x-axis.
    - y_col: str
        The name of the column to be used as the y-axis.
    - hue_col: str (optional)
        The name of the column to be used for marker colors.

    Returns:
    - Tuple[plt.Figure, plt.Axes]
        Returns a tuple containing the Figure and Axes objects for the scatterplot.
    '''
    # Create the scatterplot
    fig, ax = plt.subplots(figsize=(10, 6))

    if hue_col is None:
        fig=sns.scatterplot(data=df, x=x_col, y=y_col, ax=ax)
    else:
        fig=sns.scatterplot(data=df, x=x_col, y=y_col, hue=hue_col, palette='Set2', ax=ax)

    # Set plot title and labels
    ax.set_title(f'Scatter Plot: {x_col} vs {y_col}')
    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)

    # Add legend
    plt.legend(title=hue_col, loc='upper right')

    # Show the plot
    plt.show()

    return fig, ax
```
